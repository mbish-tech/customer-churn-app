```python
import pandas as pd
import numpy as np
import random
from faker import Faker

# Initialize faker and random seed
fake = Faker()
np.random.seed(42)
random.seed(42)

# Set dataset size
n_customers = 1000

# Options for categorical variables
subscription_plans = ['Basic', 'Standard', 'Premium']
payment_methods = ['Credit Card', 'PayPal', 'Mobile Money']

data = []

for i in range(n_customers):
    customer_id = fake.uuid4()
    gender = random.choice(['Male', 'Female'])
    age = np.random.randint(18, 70)
    subscription_plan = random.choices(subscription_plans, weights=[0.5, 0.3, 0.2])[0]
    
    # Monthly fees based on plan
    if subscription_plan == 'Basic':
        monthly_fee = 10
    elif subscription_plan == 'Standard':
        monthly_fee = 20
    else:
        monthly_fee = 30
    
    total_watch_hours = round(np.random.exponential(scale=20), 1)  # watch hours skewed
    last_login = np.random.randint(0, 30)
    payment_method = random.choice(payment_methods)
    support_tickets = np.random.poisson(0.5)
    
    # Churn logic (simulate some pattern)
    churn_probability = 0.1 + (last_login / 60) + (support_tickets / 10) - (total_watch_hours / 200)
    churn_probability = min(max(churn_probability, 0), 1)  # clamp between 0 and 1
    churn = np.random.choice([1, 0], p=[churn_probability, 1 - churn_probability])
    
    data.append({
        'CustomerID': customer_id,
        'Gender': gender,
        'Age': age,
        'SubscriptionPlan': subscription_plan,
        'MonthlyFee': monthly_fee,
        'TotalWatchHours': total_watch_hours,
        'LastLogin': last_login,
        'PaymentMethod': payment_method,
        'SupportTickets': support_tickets,
        'Churn': churn
    })

# Create dataframe
df = pd.DataFrame(data)

# Export to CSV (optional)
df.to_csv('customer_churn.csv', index=False)

# Preview
df.head()
```

```python
# 1️⃣ Import libraries
import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE

# 2️⃣ Load your dataset
df = pd.read_csv('customer_churn.csv')

# 3️⃣ Creating new engineered features
df['WatchHoursPerDay'] = df['TotalWatchHours'] / 30
df['HeavyUser'] = (df['TotalWatchHours'] > 30).astype(int)
df['RecentLogin'] = (df['LastLogin'] <= 7).astype(int)
df['Support_Watch_Interaction'] = df['SupportTickets'] * (30 - df['TotalWatchHours'])
df['LoginBucket'] = pd.cut(df['LastLogin'], bins=[-1, 7, 14, 30], labels=['0-7', '8-14', '15-30'])

# 4️⃣ Encode categorical variables
df_encoded = pd.get_dummies(df, columns=['Gender', 'SubscriptionPlan', 'PaymentMethod', 'LoginBucket'], drop_first=True)

# 5️⃣ Define features and target
X = df_encoded.drop(['CustomerID', 'Churn'], axis=1)
y = df_encoded['Churn']

# 6️⃣ Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 7️⃣ Apply SMOTE to handle imbalance
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Check class distribution after SMOTE
print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", pd.Series(y_train_res).value_counts())

# 8️⃣ Train Random Forest Model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_res, y_train_res)

# 9️⃣ Predict
y_pred = model.predict(X_test)

# 🔟 Evaluate
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("AUC Score:", roc_auc_score(y_test, model.predict_proba(X_test)[:,1]))

# 🔥 BONUS: Plot feature importances
importances = model.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(y='Feature', x='Importance', data=feature_importance_df)
plt.title('Feature Importances')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create model
rf = RandomForestClassifier(random_state=42)

# Create grid search
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=5, n_jobs=-1, scoring='roc_auc', verbose=2)

# Fit on resampled data
grid_search.fit(X_train_res, y_train_res)

# Best params
print("Best parameters:", grid_search.best_params_)

# Evaluate best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("AUC Score:", roc_auc_score(y_test, best_model.predict_proba(X_test)[:,1]))
```

```python
import xgboost as xgb

# Create XGBoost model
xgb_model = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    scale_pos_weight= (len(y_train_res) - sum(y_train_res)) / sum(y_train_res)  # to handle imbalance
)

# Train on resampled data
xgb_model.fit(X_train_res, y_train_res)

# Predict
y_pred_xgb = xgb_model.predict(X_test)

# Evaluate
print(confusion_matrix(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
print("AUC Score:", roc_auc_score(y_test, xgb_model.predict_proba(X_test)[:,1]))
```
